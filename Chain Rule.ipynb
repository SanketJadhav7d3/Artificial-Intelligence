```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def deriv(func, input_, delta=0.001):
    '''
        Derivative of function
    '''
    return (func(input_ + delta) - func(input_ - delta)) / (2 * delta )
```

```python
def sigmoid(x):
    return np.square(1 / (1 + np.exp(-x)))
```

```python
input_ = np.arange(-5, 5, 0.1)
```

```python
%matplotlib inline
plt.plot(input_, sigmoid(input_), label="sigmoid")
plt.plot(input_, deriv(sigmoid, input_), label="derivative of sigmoid") 
plt.legend()
```

```python
 %matplotlib inline
plt.plot(input_, np.tanh(input_), label="tanh")
plt.plot(input_, deriv(np.tanh, input_), label="derivative of tanh") 
plt.legend()
```
